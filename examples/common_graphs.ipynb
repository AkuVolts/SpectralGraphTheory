## Some Example Graphs with their Eigenvalues and Eigenvectors

```julia
using Graphs
using GraphPlot
using LinearAlgebra
using Plots
```

### 1. Paths

```julia
# make a path with 4 vertices
g = path_graph(4)

# plot the graph
gplot(g, nodelabel=1:4)
```

```julia
# make a matrix of the graph
A = Matrix(adjacency_matrix(g))
```

```julia
# get the laplacian matrix
L = Matrix(laplacian_matrix(g))
```

```julia
# get the eigenvalues of the laplacian matrix
E = eigvals(L)
Evecs = eigvecs(L)
println(E)
```

```julia
# Initialize the first plot
plot(Evecs[:, 1], marker=5, label="Eigenvector 1", legend=:false)

# Add the remaining eigenvectors to the same plot
for i in 2:4
    plot!(Evecs[:, i], marker=5, label="Eigenvector $i")
end

# Add axis labels
xlabel!("vertex number")
ylabel!("value in eigenvector")
```

```julia
# do the same thing with more vertices
g = path_graph(10)

# plot the graph
gplot(g, nodelabel=1:10)

# make a matrix of the graph
A = Matrix(adjacency_matrix(g))

# get the laplacian matrix
L = Matrix(laplacian_matrix(g))

# get the eigenvalues of the laplacian matrix
E = eigvals(L)
Evecs = eigvecs(L)

# print the eigenvalues
println(E)
println(Evecs) 
```

```julia
# plot all the vectors in the eigenvector matrix 
for i in 1:10
    p = plot(Evecs[:, i],marker=5,legend=true, label="eigenvector $i")
    xlabel!("vertex number")
    ylabel!("value in eigenvector")
    display(p)
end

```

```julia

# Initialize the first plot
plot(Evecs[:, 1], marker=5, label="Eigenvector 1", legend=:false)

# Add the remaining eigenvectors to the same plot
for i in 2:10
    plot!(Evecs[:, i], marker=5, label="Eigenvector $i")
end

# Add axis labels
xlabel!("vertex number")
ylabel!("value in eigenvector")
```

### 2. Cycles

```julia
g_cycle = cycle_graph(10)

# plot the graph
gplot(g_cycle, nodelabel=1:10)
```

```julia
# make a matrix of the graph
A = Matrix(adjacency_matrix(g_cycle))
println(A)

# get the laplacian matrix
L = Matrix(laplacian_matrix(g_cycle))
println(L)

# get the eigenvalues of the laplacian matrix
E = eigvals(L)
Evecs = eigvecs(L)

# print the eigenvalues
println(E)
println(Evecs) 
```

```julia
# plot all the vectors in the eigenvector matrix 
for i in 1:10
    p = plot(Evecs[:, i],marker=5,legend=true, label="eigenvector $i")
    xlabel!("vertex number")
    ylabel!("value in eigenvector")
    display(p)
end
```

```julia
# Initialize the first plot
plot(Evecs[:, 1], marker=5, label="Eigenvector 1", legend=:false)

# Add the remaining eigenvectors to the same plot
for i in 2:10
    plot!(Evecs[:, i], marker=5, label="Eigenvector $i")
end

# Add axis labels
xlabel!("vertex number")
ylabel!("value in eigenvector")
```
